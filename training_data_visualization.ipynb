```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy.stats import linregress

elevation_data = data_frame.dropna(subset=['Elevation Gain (m)'])
elevation_data['Start Time'] = pd.to_datetime(elevation_data['Start Time'], utc=True)
elevation_data.set_index('Start Time', inplace=True)

weekly_elevation_data = elevation_data.resample('W').sum()
weekly_elevation_data = weekly_elevation_data.dropna()
weekly_elevation_data = weekly_elevation_data.reset_index()

dates = np.array(weekly_elevation_data['Start Time']).astype(np.datetime64)
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_format = mdates.DateFormatter('%Y')

weekly_elevation_slope, weekly_elevation_intercept, _, _, _ = linregress(
    np.arange(len(dates)), weekly_elevation_data['Elevation Gain (m)']
)

_, ax = plt.subplots()

ax.set_title('Weekly Elevation Gained In Meters Over Time')
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(years_format)
ax.xaxis.set_minor_locator(months)

ax.plot(dates, weekly_elevation_slope * np.arange(len(dates)) + weekly_elevation_intercept, label='Average Elevation Gain (m)')
ax.plot(dates, weekly_elevation_data['Elevation Gain (m)'], label='Elevation Gain (m)')
ax.legend(loc='best')
```
